# Does self-reporting penis size inflate the measurement results? 
### Is there a relation between penis size and happiness?
In order to conduct this "study" I am going to use [Kaggle](https://www.kaggle.com) to get my [Penis Size](https://www.kaggle.com/datasets/antimoni/international-study-on-male-genital-measurements/) dataset and the [2019 World Happiness Report](https://www.kaggle.com/datasets/unsdsn/world-happiness) dataset.

```python
# Load the CSV files and libraries
import pandas as pd
import matplotlib.pyplot as plt

penis_data = pd.read_csv('penis.csv')
happiness_data = pd.read_csv('happiness.csv')
```

```python
# Merge both data frames by Country name
data = pd.merge(penis_data, happiness_data, left_on='Country', right_on='Country or region', how='left')

# Remove entires that don't have all the data needed
data.dropna(how='any', inplace=True)
```

Now let's eyeball the data to decide which approach to take, current relations of interest are:
- Penis Size self-reported x measured (global and regionally)
- Penis Size x Happiness

```python
# Penis Size self-reported x measured (global)
plt.xlabel('Length Erect')
plt.ylabel('Length Growth')
groups = data.groupby('Method')
for name, group in groups:
    plt.scatter(group.length_erect, group.growth_length, label=name, s=5)
plt.legend()
```

At a first glance we can detect some interesting observations:
- There seem to be 4 categories of penises in this relation, categorization should be interesting to do.
- The *Self reported* sizes seem to be bigger than the *Measured* sizes on average. Multiple causes might cause this such as access to doctors, weather and most probably, pride.
- For the same region of *Growth* - besides the same observation as seen before - we can see that for 3 of the 4categories *Erect Size* and *Growth* are sort of related in a ascending order, with the exception of 1 of the categories that shares a similar amount of *Erect Length* with a small *Growth*, no predictions about this phenomenon so far, further study should yield interesting results.
